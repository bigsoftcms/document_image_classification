## 7 Topics: Overlap of Topics 2 and 7 | Non-Uniformity in Topic Sizes

```python
import gensim
import pyLDAvis.gensim
```

```python
# load dictionary, bow_corpus, and lda model
dictionary = gensim.corpora.Dictionary.load('LDAvis_choose_n_topics/dict_7.dict')
corpus = gensim.corpora.MmCorpus('LDAvis_choose_n_topics/bow_corpus_7.mm')
lda = gensim.models.ldamodel.LdaModel.load('LDAvis_choose_n_topics/lda_7.model')

# note: topic numbers here are not the same as assigned by gensim.
# these topic numbers are assigned in descending order corresponding to topic with largest percentage of tokens from entire corpus.
to_disp = pyLDAvis.gensim.prepare(lda, corpus, dictionary)
pyLDAvis.display(to_disp)
```
